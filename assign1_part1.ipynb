```python
import pandas as pd
import numpy as np
import matplotlib as mpe
import seaborn as sns
%matplotlib inline
```

```python
country = pd.read_excel('countries_regions.xlsx')
country.head()
```

# GET DATA Broadband

```python
broadband1 = pd.read_excel('infrastructure/broadband.xlsx')
broadband1.replace({'..':np.nan},inplace=True)
broadband1.head()
```

```python
broadband1.keys()
```

```python
broadband1 = country.join(broadband1[['2000 [YR2000]', '2014 [YR2014]']])
broadband1.head()
```

```python
broadband1 = broadband1[broadband1.Region != "Thailand"]
broadband1 = broadband1[broadband1.Region != "Australia and Oceanea"]
```

```python
sns.heatmap(broadband1.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
broadband1.groupby("Region")["2000 [YR2000]"].describe()
```

```python
broadband1.groupby("Region")["2014 [YR2014]"].describe()
```

```python
broadband1 = broadband1.dropna()
broadband1.head()
```

```python
#broadband with outliers
```

```python
chart = sns.boxplot(x="Region",y="2000 [YR2000]",data=broadband1)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
fig = chart.get_figure()
fig.savefig('pic_boxplot_part1/boxplot_broadband2000_with_outliers.jpg', orientation="horizontal", dpi = 300)
```

```python
chart = sns.boxplot(x="Region",y="2014 [YR2014]",data=broadband1)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
fig = chart.get_figure()
fig.savefig('pic_boxplot_part1/boxplot_broadband2014_with_outliers.jpg', orientation="horizontal", dpi = 300)
```

```python
#broadband without outliers
```

```python
chart = sns.boxplot(x="Region",y="2000 [YR2000]",data=broadband1, showfliers = False)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
fig = chart.get_figure()
fig.savefig('pic_boxplot_part1/boxplot_broadband2000_without_outliers.jpg', orientation="horizontal", dpi = 300)
```

```python
chart = sns.boxplot(x="Region",y="2014 [YR2014]",data=broadband1, showfliers = False)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
fig = chart.get_figure()
fig.savefig('pic_boxplot_part1/boxplot_broadband2014_without_outliers.jpg', orientation="horizontal", dpi = 300)
```
